##### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

**Matplotlib** is a popular open-source Python library used for creating static, interactive, and animated visualizations in a wide range of formats. It provides a wide variety of tools for producing high-quality graphs and plots, making it a go-to choice for data visualization tasks in scientific research, data analysis, and more.

Matplotlib is used for various purposes, including:

1. **Data Visualization**: Matplotlib is primarily used to create various types of visualizations to represent data in a graphical format, which makes it easier to understand and interpret complex patterns and relationships within the data.

2. **Exploratory Data Analysis**: It is extensively used in data analysis workflows to explore and understand the underlying structure and distribution of the data before performing more complex analyses.

3. **Report Generation**: Matplotlib allows users to generate plots and figures that can be incorporated into reports, presentations, scientific papers, and other documents.

4. **Model Evaluation and Validation**: Visualizing model outputs, predictions, and evaluation metrics can provide insights into the performance of machine learning models, aiding in their development and refinement.

5. **Educational and Research Purposes**: Matplotlib is commonly used in educational settings to teach data visualization concepts and in research to present findings visually.

The **'pyplot'** module of Matplotlib provides a simple and intuitive interface for creating a wide variety of plots. Here are five types of plots that can be created using the **'pyplot'** module:

1. **Line Plot**: A line plot is used to visualize the relationship between two continuous variables. It's often used to show trends over time or correlations between variables.

2. **Scatter Plot**: A scatter plot displays individual data points as dots on a two-dimensional plane, useful for visualizing the relationship between two variables and identifying patterns.

3. **Bar Plot**: A bar plot (bar chart) is used to compare different categories or groups by representing the data using vertical or horizontal bars. It's suitable for categorical data.

4. **Histogram**: A histogram is used to represent the distribution of a continuous variable by dividing it into bins and displaying the frequency of data points in each bin.

5. **Pie Chart**: A pie chart is used to display the proportions of different categories as slices of a circular chart. It's suitable for representing parts of a whole.

##### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

**Note:** Also add title, xlabel, and ylabel to the plot.

A **scatter plot** is a type of data visualization that uses dots to represent individual data points in a two-dimensional space. It's particularly useful for visualizing the relationship between two continuous variables and identifying patterns, trends, clusters, or correlations in the data.

In a scatter plot, each dot represents a single data point, where the horizontal position of the dot corresponds to the value of one variable, and the vertical position corresponds to the value of the other variable. This arrangement allows you to observe how changes in one variable are related to changes in the other variable.

Key characteristics of a scatter plot include:

1. **X-axis and Y-axis**: The two axes represent the two variables being compared. The variable represented by the X-axis is often referred to as the independent variable, while the one represented by the Y-axis is the dependent variable.

2. **Dots**: Each data point is represented by a dot on the plot. The position of the dot reflects the values of the two variables it represents.

3. **Patterns and Relationships**: Scatter plots are used to identify patterns in the data. These patterns can include positive or negative correlations (when one variable increases, the other tends to increase or decrease), clusters (groups of data points close to each other), or lack of correlation (random scattering).

4. **Title and Labels**: A scatter plot should have a title that describes the content of the plot. Additionally, labels for the X-axis and Y-axis should be provided to indicate the variables being represented.

5. **Legend**: If multiple groups or categories are represented in the scatter plot, a legend can be included to provide context for different colors or shapes used for different groups.

Scatter plots are an effective way to visually explore the relationships between two variables, which can help in making informed decisions, identifying outliers or anomalies, and forming hypotheses about underlying patterns in the data. They are widely used in various fields including statistics, data analysis, machine learning, and scientific research.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
```

##### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,1,2,3,4,5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('Line 1')

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('Line 2')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('Line 3')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('Line 4')

plt.tight_layout()
```

##### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A **bar plot**, also known as a bar chart, is a data visualization that uses rectangular bars to represent data values. Each bar's length or height corresponds to the value of a particular variable or category. Bar plots are commonly used to display and compare the frequency, count, or other measures associated with different categories or groups.

Key features of bar plots include:

1. **Categorical Data Representation**: Bar plots are particularly useful when you have categorical data, where each category is represented by a bar. Categories can represent anything from different groups, classes, or labels to specific time periods, regions, or any other discrete classification.

2. **Comparisons**: Bar plots are effective for comparing values among different categories. They allow you to easily identify the largest or smallest values, as well as the relative sizes of values across categories.

3. **Labels**: Each bar is often labeled with the corresponding category name or label, making it easy to interpret the data.

4. **Orientation**: Bar plots can be created with either vertical bars (where the length of the bar represents the value) or horizontal bars (where the height of the bar represents the value). The choice of orientation depends on the data and the visual presentation you want to achieve.

5. **Bar Types**: There are variations of bar plots, such as grouped bar plots, stacked bar plots, and percentage bar plots, each suited for different types of data and comparison scenarios.

Bar plots are used for various purposes:

- **Comparison**: They are excellent for visualizing the differences in data among different categories. For example, you can use a bar plot to compare sales performance across different products or the population distribution among different age groups.

- **Distribution**: Bar plots can also show the distribution of data within a single category. For instance, a histogram is a type of bar plot that displays the frequency of data points within different bins.

- **Trends**: Bar plots can be used to represent trends over time or across different periods, such as monthly revenue or quarterly sales figures.

- **Highlighting Patterns**: By using different colors, patterns, or shading, you can highlight specific patterns or characteristics within the data, making the visualization more informative.

- **Presentation**: Bar plots are commonly used in presentations, reports, and articles to communicate data insights to a broader audience.

Overall, bar plots are a straightforward and powerful tool for visually summarizing and communicating information about categorical data, making them a staple in the field of data visualization and analysis.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
plt.bar(company, profit)
plt.title('Bar plot')

plt.subplot(2,2,2)
plt.barh(company, profit)
plt.title('Horizontal Bar plot')

plt.tight_layout()
```

##### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A **box plot**, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the distribution of a dataset. It displays key statistical measures, such as the median, quartiles, and potential outliers, using a simple graphical representation. Box plots are particularly useful for identifying the central tendency, spread, and potential skewness or asymmetry in the data.

Here's how a box plot is constructed:

- **Box**: The central rectangle, or "box," represents the interquartile range (IQR), which is the range between the first quartile (25th percentile) and the third quartile (75th percentile) of the data. This box spans the middle 50% of the data.

- **Median Line**: A line is drawn within the box to represent the median (50th percentile) of the data. The median divides the data into two halves, with 50% of the data points lying below it and 50% above it.

- **Whiskers**: Lines, or "whiskers," extend from the edges of the box to the minimum and maximum values within a certain range. The exact range of the whiskers can vary based on the data distribution and any potential outliers.

- **Outliers**: Data points that fall significantly outside the whiskers' range are considered outliers and are often plotted individually using dots or other symbols. These outliers can provide insights into potential anomalies or unusual data points.

Box plots are used for various purposes:

1. **Visualizing Distribution**: Box plots provide a concise visual summary of the data distribution, showing the spread, central tendency, and presence of outliers in a single plot.

2. **Comparison**: Box plots are effective for comparing the distributions of multiple datasets or groups. By placing box plots side by side, you can easily compare the central tendencies, spreads, and shapes of the distributions.

3. **Identifying Outliers**: Box plots help in identifying potential outliers that might indicate data anomalies, errors, or interesting data points that deserve further investigation.

4. **Data Skewness**: The symmetry or skewness of the distribution can be visually assessed from the length of the whiskers on each side of the box.

5. **Understanding Spread**: The IQR represented by the box provides insights into the range within which the middle 50% of the data points lie, helping to understand data variability.

6. **Presentation**: Box plots are often used in reports, presentations, and research papers to provide a clear and intuitive representation of the data distribution.

Overall, box plots are valuable tools for exploratory data analysis, making informed comparisons between datasets, and gaining insights into the distribution and characteristics of your data.

```python
import numpy as np
import matplotlib.pyplot as plot

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
plt.boxplot(box1)
plt.title('Box 1')

plt.subplot(2,2,2)
plt.boxplot(box2)
plt.title('Box 2')

plt.tight_layout()
```
